```python
from IPython.display import clear_output
from time import sleep
import random
import sqlite3

con = sqlite3.connect('CAH.db') # connects to the database
cur = con.cursor()

def Game(NumberPlayers):
    Czar = NumberPlayers - 1
    NameDB = "QuestionCards" 
    NumPoints = []
    Continue = True
    Winner = -1
    WinningNumPoints = 2
    PlayerInfo = GiveCardsPlayers(NumPlayers)
    for i in range(NumberPlayers):
        NumPoints.append(0)
    while Continue:
        QuestionCard, QuestionID = GetCard(NameDB)
        NumGaps = GetNumGaps(QuestionID)
        ListOfChoices = []
        CzarChoice = 0
        print(f"the Czar this round is Player {Czar + 1}")
        for index2 in range(NumberPlayers):
            input(f"Player {index2 + 1} press enter when ready")
            if index2 == Czar:
                continue
            else:
                Choice = []
                Answer = []
                Choice.append(index2)
                print(f"""
Player {index2 + 1} Turn
                
Question Card:
{QuestionCard}

Your Answer Cards:
{PlayerInfo[index2][1]}""")
                for Gaps in range(NumGaps):                    
                    num = int(input(f"What is your {Gaps + 1} choice (1 - 7): "))
                    Answer.append(PlayerInfo[index2][1][num - 1])
                    
                Choice.append(Answer)
                ListOfChoices.append(Choice)
                
            sleep(2)
            clear_output()
        print(f"""
Player {index2 + 1} Turn
                
Question Card:
{QuestionCard}
                
Submitted Answer Cards:""")
        for index3 in range(len(ListOfChoices)):
            print(ListOfChoices[index3][1])
        CzarChoice = int(input(f"What is your choice (1 - {NumberPlayers - 1})? "))
        PointWinner = ListOfChoices[CzarChoice - 1][0]
        NumPoints[PointWinner] += 1
        if NumPoints[PointWinner] == WinningNumPoints:
            Continue = False
            Winner = PointWinner
        if Czar != 0:
            Czar -= 1
        else:
            Czar = NumberPlayers - 1
        print(f"Congratulations Player {PointWinner + 1} for winning this round")
        print("The points this round are:")
        for i in range(len(NumPoints)):
            print(f"Player {i + 1} => {NumPoints[i]}")
    print(f"Congratulations the winner is Player {Winner + 1}")
            
def GetCard(Type):
    MaxCardId = cur.execute(f"SELECT COUNT() FROM {Type}").fetchone()
    RandomNum = random.randint(1, MaxCardId[0])
    for card in cur.execute(f"SELECT card FROM {Type} WHERE id = {RandomNum}"):
        break
    return card[0], RandomNum
    
def CreatePlayerHand(NumCards):
    CardType = "AnswerCards"
    list = []
    for index in range(NumCards):
        Card, ids = GetCard(CardType)
        list.append(Card)
    return list

def GiveCardsPlayers(NumPlayers):
    NumCardsHand = 7
    PlayerInfo = []
    Info = []
    for index in range(NumPlayers):
        Info.append(index + 1)
        Info.append(CreatePlayerHand(NumCardsHand))
        PlayerInfo.append(Info)
        Info = []
    return PlayerInfo

def GetNumGaps(QCardID):
    for Gaps in cur.execute(f"SELECT NumGaps FROM QuestionCards WHERE id = {QCardID}"):
        break
    return Gaps[0]
    
NumPlayers = 0
CardsHand = []

NumPlayers = int(input("How many players? "))
Game(NumPlayers)

con.close()
```

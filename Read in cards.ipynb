```python
def ReadInFile(FileName):
    """
    This function reads in the file and saves each line of the file into a list and returns that list
    """
    with open(FileName, "r", encoding="utf-8") as file: # opens the csv file as file
        lines = file.readlines() # saves each line of the file into the list lines
    return lines
```

```python
def ReturnCards(StartLine, EndLine, lines):
    """
    this function goes through the list and recognises each card and saves them in a list and returns that list
    """
    # Initilasing variables
    cards = []
    card = ""
    # for loop that goes through the cards and saves them in a list
    for lineNum in range(StartLine, EndLine):
        # each card is seperated by a line, so this if checks if the line is a part of the old card or the start of a new one
        if lines[lineNum] != "\n":
            # adds the line to the variable card and adds a space to make the card look neater
            card += lines[lineNum].strip()
            card += " "
        else:
            cards.append(card.strip()) # adds the finished card to the list
            card = "" # the card is wiped after appending for the start of another card
    return cards
```

```python
def GapReplace(Card):
    """
    This function replaces the gap of a card and returns the edited card and the number of gaps it has
    """
    LocationOfGap = 0 # variable that will hold the location of the start of the gap
    EndOfGap = 0 # location of the end of the gap
    NumOfGaps = 0 # variable that will contain the number of gaps, if any, of the card
    count = 0
    
    while "_" in Card: # a while loop that will loop until there are no more gaps in the string
        LocationOfGap = Card.find("_") # finds the location of the first gap
        while (LocationOfGap + count) < len(Card) and Card[LocationOfGap + count] == "_": # while loop that counts how many _ there are
            count += 1
        EndOfGap = count + LocationOfGap # calculates the location of the end of the gap
        Card = Card[:LocationOfGap] + "{}" + Card[(EndOfGap):] # replaces the gaps with {}
        count = 0
        NumOfGaps += 1 # adds to the number of gaps
        
    if NumOfGaps == 0: # an if statement to catch a card that has no gaps in itand adding a {} to the end
        Card += " {}" # adds {x} to the end of the card
        NumOfGaps = 1
        
    return Card, NumOfGaps
```

### This is the part where the lines of the file are printed out and the functions are ran

```python
CAHFileName = "CAH_MainGame.txt" # Run this if you want to play the X rated game
```

```python
CAHFileName = "CAH_Family_SmallCards.txt" # Run this if you want to play the R rated game
```

```python
# prints out the lines of the pdf in a format which has the line number first
lines = ReadInFile(CAHFileName) # reads in the file lines and puts them in a lsit
for i, line in enumerate(lines): # goes thorugh all the lines and makes i = line number and line = the line
    print(f"{i}: {line.strip()}") # prints put the line in the form "1: Line1"
```

```python
if CAHFileName == "CAH_MainGame.txt":
    # Inititalising variables
    AnswerLineStart = 133 # the line number in which the answer cards start
    AnswerLineEnd = 1524 # the line number in which the end of the last answer card is located
    QuestionLineStart = 1524 # the line number in which the question cards start
    QuestionLineEnd = 1919 # the line number in which the end of the last question card is located
elif CAHFileName == "CAH_Family_SmallCards.txt":
    # Inititalising variables
    AnswerLineStart = 551 # the line number in which the answer cards start
    AnswerLineEnd = 1877 # the line number in which the end of the last answer card is located
    QuestionLineStart = 65 # the line number in which the question cards start
    QuestionLineEnd = 551 # the line number in which the end of the last question card is located
```

```python
import string # imports tring to be able to access the constant punctuation to be able to strip the answer cards

FinishedQCards = []

# calling functions
ListOfLines = ReadInFile(CAHFileName) # makes ListOfLines equal a list of all the lines of the file
AnswerCards = ReturnCards(AnswerLineStart, AnswerLineEnd, ListOfLines) # makes the AnswerCards equal a list of all the answer cards
QuestionCards = ReturnCards(QuestionLineStart, QuestionLineEnd, ListOfLines) # makes the QuestionCards equal a list of all the question cards

# for loop to go thorugh all the cards and strip all the punctuation such as ? , . ! at the end of the of each card
for index in range(len(AnswerCards) - 1):
    AnswerCards[index] = AnswerCards[index].strip(string.punctuation)
    AnswerCards[index] = AnswerCards[index].strip()

# a for loop to go through all the question cards and change them
for index in range(len(QuestionCards) - 1):
    QCard, NumOfGaps = GapReplace(QuestionCards[index])
    # saves all the cards and number of gaps in a tuples inside the list
    FinishedQCards.append((QCard, NumOfGaps))
```

### Saves the cards in the database

```python
import sqlite3

con = sqlite3.connect('CAH.db') # connects to the database
cur = con.cursor() # creates a cursor
```

```python
# checks to see if there is already a table called AnswerCards or QuestionCards and if there is it deletes it
try:
    cur.execute("DROP TABLE AnswerCards")
    cur.execute("DROP TABLE QuestionCards")
    cur.execute("DROP TABLE Players")
    cur.execute("DROP TABLE Games")
except:
    pass

# creates the table AnswerCards for the Answer cards with the properties: id, card(the actual card)
cur.execute("""CREATE TABLE AnswerCards 
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                card text)""")

# creates the table QuestionCards for the Question cards with the properties: id, card(the actual card), NumGaps (the number of gaps in the card)
cur.execute("""CREATE TABLE QuestionCards 
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                card text,
                NumGaps int)""")

# creates the table Players to keep track of the players info with the properties: id, ListCards (a JSON string that represents the list of cards), choice, points
cur.execute("""CREATE TABLE Players 
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                ListCards text,
                choice text,
                points int)""")

# creates the table Games to keep track of the game info with the properties: id, ListPlayerID (a JSON string that represents the list of ids), CzarID
cur.execute("""CREATE TABLE Games 
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                ListPlayerID text,
                QCard text,
                CzarID int,
                EveryoneIn Boolean)""")

# loops through all the question cards and saves them in the table with an id added to it
for index in range(len(FinishedQCards) - 1):
    cur.execute(f"""INSERT INTO QuestionCards (card, NumGaps) VALUES ("{FinishedQCards[index][0]}", {FinishedQCards[index][1]})""")

# loops through all the answer cards and saves them in the table with an id added to it
for index in range(len(AnswerCards) - 1):
    cur.execute(f"""INSERT INTO AnswerCards (card) VALUES 
                    ("{AnswerCards[index]}")""")
    
# commits the table into the database
con.commit()
```

```python
# prints out the whole table
for card in cur.execute("SELECT * FROM QuestionCards"):
    print(card)

# prints out the whole table
for card in cur.execute("SELECT * FROM AnswerCards"):
    print(card)
```

```python
con.close() # closes the database and saves it
```
